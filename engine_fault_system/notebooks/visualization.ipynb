```python
import pandas as pd

# Step 2: Load Dataset
data = pd.read_csv("C:\\Users\\ADMIN\\Desktop\\CI LAB\\engine_fault_system\\data\\engine_data.csv")
print("✅ Engine Data Loaded:", data.shape)
print(data.head())
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# -------------------------------
# 1️⃣ Correlation Heatmap
# -------------------------------
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Heatmap of Engine Parameters")
plt.tight_layout()
plt.show()

# -------------------------------
# 2️⃣ Pairplot — parameter relations
# -------------------------------
sns.pairplot(data, hue="Engine Condition", diag_kind='kde', palette="coolwarm")
plt.suptitle("Pairwise Relationship of Engine Parameters", y=1.02)
plt.show()

# -------------------------------
# 3️⃣ Boxplots — for outlier detection
# -------------------------------
num_cols = ['Engine rpm', 'Lub oil pressure', 'Fuel pressure', 
            'Coolant pressure', 'lub oil temp', 'Coolant temp']

plt.figure(figsize=(10, 6))
data[num_cols].boxplot()
plt.title("Boxplot of Engine Sensor Readings")
plt.ylabel("Sensor Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# 4️⃣ Condition Distribution
# -------------------------------
plt.figure(figsize=(6, 4))
sns.countplot(x="Engine Condition", data=data, palette="Set2")
plt.title("Distribution of Engine Conditions")
plt.xlabel("Condition (0 = Normal, 1 = Faulty)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\CI LAB\engine_fault_system\data\vehicle_emission_dataset_synthetic_v3_labeled.csv')

# -------------------------------
# Correlation Heatmap
# -------------------------------
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlOrBr", fmt=".2f")
plt.title("Correlation Heatmap — Vehicle Emission Parameters")
plt.tight_layout()
plt.show()

# Boxplot 1: CO2 Emissions
# -------------------------------
plt.figure(figsize=(6, 5))
sns.boxplot(y=df["CO2 Emissions"], color="lightcoral", width=0.4)
plt.title("CO₂ Emissions Distribution")
plt.ylabel("CO₂ (g/km)")
plt.tight_layout()
plt.show()

# -------------------------------
# Boxplot 2: Other Emissions (NOx, PM2.5, VOC, SO2)
# -------------------------------
other_cols = ['NOx Emissions', 'PM2.5 Emissions', 'VOC Emissions', 'SO2 Emissions']

plt.figure(figsize=(8, 5))
sns.boxplot(data=df[other_cols], palette="Set2")
plt.title("Distribution of Other Emissions")
plt.ylabel("Emission (g/km)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
